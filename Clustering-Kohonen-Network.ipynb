```python
import numpy as np
import pandas as pd
```

```python
names = ['usr','q4_sent','q4_received','q4_access','q4_comments_on_sent','q4_comments_on_received','q4_likes_on_sent','q4_likes_on_received','working_days']
ds = pd.read_csv('c:/ariel/subset_globant_features.csv', names=names)
```

```python
ds.head()
```

```python
#normalization

list_sl=[]
list_sw=[]
list_pl=[]
list_pw=[]
for sl in ds['q4_sent']:
    sl = (sl-min(ds['q4_sent']))/(max(ds['q4_sent'])-min(ds['q4_sent']))
    list_sl.append(sl)
for sw in ds['q4_received']:
    sw = (sw-min(ds['q4_received']))/(max(ds['q4_received'])-min(ds['q4_received']))
    list_sw.append(sw)    
for pl in ds['q4_access']:
    pl = (pl-min(ds['q4_access']))/(max(ds['q4_access'])-min(ds['q4_access']))
    list_pl.append(pl)
for pw in ds['q4_comments_on_sent']:
    pw = (pw-min(ds['q4_comments_on_sent']))/(max(ds['q4_comments_on_sent'])-min(ds['q4_comments_on_sent']))
    list_pw.append(pw) 

X = np.array( list(zip(list_sl,list_sw, list_pl, list_pw)) )
```

```python
print (X)
```

```python
nc = 3         # number of classes
W = []         # list for w vectors
M = len(X)     # number of x vectors
N = len(X[0])  # dimensionality of x vectors
```

```python
def get_weights():
    y = np.random.random() * (2.0 / np.sqrt(M))
    return 0.5 - (1 / np.sqrt(M)) + y
```

```python
for i in range(nc):
    W.append(list())
    for j in range(N):
        W[i].append(get_weights() * 0.5)
```

```python
def distance(w, x):
    r = 0
    for i in range(len(w)):
        r = r + (w[i] - x[i])*(w[i] - x[i])
    
    r = np.sqrt(r)
    return r
```

```python
def Findclosest(W, x):
    wm = W[0]
    r = distance(wm, x)
    
    i = 0
    i_n = i
    
    for w in W:
        if distance(w, x) < r:
            r = distance(w, x)
            wm = w
            i_n = i
        i = i + 1
    
    return (wm, i_n)
```

```python
print(W)
```

```python
la = 0.3    # λ coefficient
dla = 0.05  # Δλ

while la >= 0:
    for k in range(10):
        for x in X:
            wm = Findclosest(W, x)[0]
            for i in range(len(wm)):
                wm[i] = wm[i] + la * (x[i] - wm[i]) 

    la = la - dla
```

```python
Data = list() 

for i in range(len(W)):
    Data.append(list())

dfList = ds['usr'].as_matrix()
```

```python
DS = list()
i = 0
for x in X:
    i_n = Findclosest(W, x)[1]
    Data[i_n].append(x)
    DS.append([i_n, dfList[i]])
    i = i + 1

print (DS)
```
